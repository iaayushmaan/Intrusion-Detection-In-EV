## **Performance Comparison Analysis**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Comparison of Performance evaluation metrics for DoS attacks (Dataset 1)

import pandas as pd
import matplotlib.pyplot as plt

plotdata = pd.DataFrame({

    "SVM":[99, 99, 99],

    "DT":[100, 100, 100],

    "KNN":[99, 99, 99]},

    index=["Precision", "Recall", "F1 Score"])

plotdata.plot(kind="bar",figsize=(8, 6), width=0.45)

#plt.title("FIFA ratings")
#plt.legend(loc="upper left")
plt.legend(fontsize = 14, loc=(1.09, 0.05))

plt.xticks(fontsize=18, rotation=0, horizontalalignment="center")

plt.xlabel("ML Techniques", fontweight ="bold", fontsize = 20)

plt.ylabel("DoS Attack (%)", fontweight ="bold", fontsize = 20)
plt.title("DoS Performance (Dataset 1)", fontsize = 18)
plt.ylim(40, 105)


#plt.xticks([r + barWidth for r in range(len(SVM))],
        #['Precision', 'Recall', 'F1 Score'])
```
